Data Visualization II

Use the inbuilt dataset 'titanic' as used in the above problem. Plot a box plot for distribution of age with respect to each gender along with the information about whether they survived or not. (Column names : 'sex' and 'age')
Write observations on the inference from the above statistics.

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import plotly.express as px
import seaborn as sns
```

```python
temp = sns.load_dataset('titanic')
data = pd.DataFrame(temp)
data
```

```python
px.box(data,y='age',x='sex')
```

```python
sns.catplot(x='sex',hue='survived',kind='count',data=data)
```

```python
# Statistics for 'sex' column
sex_stats = data['sex'].value_counts()
# Statistics for 'age' column
age_stats = data['age'].describe()
print("Observations for 'sex'")
print(sex_stats)
print("Observations for 'age':")
print(age_stats)
```
